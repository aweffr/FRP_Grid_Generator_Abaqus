```python
# -*- coding:utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import handytools as hd
import os

a = pd.read_csv("grid.csv",header=None)
# rows = a.shape[0]
# for i in range(rows):
#     print a.loc[i].values
# for point in a.values:
#     print np.array2string(point, precision=3)
inlst = []
for point in a.values:
    inlst.append(list(point))
result = hd.coord_generator(inlst,interval=1.0, start=0.0, end=60.0)
print 'the keys of output is:\n\t',result.keys()

def result_visualize(result):
    plt.figure()
    plt.scatter(result['inCoordPoints'].getx(),result['inCoordPoints'].gety(),s=40,alpha=0.5)
    i = 0
    while True:
        point_lst = hd.mylst(result['yCoordPoints'][i:i+2])
        plt.plot(point_lst.getx(), point_lst.gety(), lw=2.5)
        i += 2
        if i + 1 > len(result['yCoordPoints']):
            break
    j = 0
    while True:
        point_lst2 = hd.mylst(result['xCoordPoints'][j:j+2])
        # pprint(point_lst2)
        plt.plot(point_lst2.getx(),point_lst2.gety(),lw=2.5)
        j += 2
        if j + 1>len(result['xCoordPoints']):
            break
    plt.show()

result_visualize(result)
hd.abaqusInputDataGenerator_shelveVersion(outPath="modelPoints-Current.dat",time=1, **result)
print "Complete!"
```
